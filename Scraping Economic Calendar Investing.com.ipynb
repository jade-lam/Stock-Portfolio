```python
import investpy as ip
```

2) Get Current Economic Calendar <br>
results are already loaded into pandas dataframe

```python
current_data = ip.economic_calendar()
current_data.info()
```

```python
current_data.head()
```

3) Scraping past economic calendar <br>
A defined date range to be imported

```python
from_date ='01/05/2022'
to_date   ='31/08/2022'

hist_data = ip.economic_calendar(
    from_date = from_date,
    to_date = to_date
)
```

```python
hist_data.info()
```
