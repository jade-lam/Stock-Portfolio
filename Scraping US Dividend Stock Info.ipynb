```python
import pandas as pd
import numpy as np
import datetime as dt
from datetime import date
import seaborn as sns
import yfinance as yf
```

```python
equity_tickers = ['ABR', 'ACRE', 'AGNC', 'ARCC', 'BST', 'EOS', 'EPD', 'HRZN', 'HTGC', 'IEP', 'LTC', 'MAIN', 'NLY', 'O', 
                  'OHI', 'ORCC', 'PDI', 'PFLT', 'PNNT', 'RC', 'RITM', 'STAG', 'STOR', 'TCPC', 'VZ', 'WPC']
etf_tickers = ['DIVO', 'JEPI', 'QYLD', 'RYLD', 'SCHD', 'XRMI', 'XYLD']
```

```python
# equity dataframe
equity_keys = ['symbol', 'longName', 'sector', 'industry', 'lastDividendDate', 'lastDividendValue', 'dividendRate', 'fiveYearAvgDividendYield',  
      'previousClose', 'twoHundredDayAverage', 'fiftyDayAverage', 'fiftyTwoWeekHigh', 'fiftyTwoWeekLow','52WeekChange']
equity_key_values = []


for ticker in equity_tickers:
  ticker_object = yf.Ticker(ticker).info
  result = [ticker_object[key] for key in equity_keys if key in ticker_object]
  equity_key_values.append(result)
```

```python
equity_df = pd.DataFrame(equity_key_values, columns=equity_keys).round(2)
equity_df.columns = ['Stock Code', 'Name', 'Sector', 'Industry', 'Last Dividend Date', 'Last Dividend (USD)', 'Forward Dividend (USD)', 
                     'Dividend Yield %', 'Previous Close (USD)', '200 day avg price (USD)', '50 day avg pricce (USD)', '52 Wk High (USD)', '52 Wk Low (USD)', '52 Wk Chg %']
equity_df['Last Dividend Date'] = pd.to_datetime(equity_df['Last Dividend Date'], unit='s', errors='coerce')
equity_df['52 Wk Chg %'] = equity_df['52 Wk Chg %']*100
```

```python
today = date.today().strftime("%B %d, %Y")
print('Date Printed:', today)
equity_df.head()
```

```python
# etf dataframe
etf_keys = ['symbol', 'longName', 'quoteType', 'fundInceptionDate', 'yield', 'threeYearAverageReturn', 'fiveYearAverageReturn', 'navPrice', 'previousClose',  
      'twoHundredDayAverage', 'fiftyDayAverage', 'fiftyTwoWeekHigh', 'fiftyTwoWeekLow']
etf_key_values = []

for etf_ticker in etf_tickers:
  etf_ticker_object = yf.Ticker(etf_ticker).info
  result = [etf_ticker_object[key] for key in etf_keys if key in etf_ticker_object]
  etf_key_values.append(result)
```

```python
etf_df = pd.DataFrame(etf_key_values, columns=etf_keys)
etf_df.columns = ['Stock Code', 'Name', 'Type', 'Fund Inception Date', 'Dividend Yield %', '3 Years Avg Return', '5 Years Avg Return', 'NAV Price (USD)', 'Previous Close (USD)', '200 day avg price (USD)', '50 day avg pricce (USD)', '52 Wk High (USD)', '52 Wk Low (USD)']
etf_df['Fund Inception Date'] = pd.to_datetime(etf_df['Fund Inception Date'], unit='s', errors='coerce')
etf_df[['Dividend Yield %', '3 Years Avg Return', '5 Years Avg Return']] = etf_df[['Dividend Yield %', '3 Years Avg Return', '5 Years Avg Return']]*100
```

```python
today = date.today().strftime("%B %d, %Y")
print('Date Printed:', today)
etf_df.round(2)
```

```python
from pandas import ExcelWriter

with pd.ExcelWriter('Dividend Yield List.xlsx') as writer:
  equity_df.to_excel(writer, sheet_name='equity list', index=False)
  etf_df.to_excel(writer, sheet_name='etf list', index=False)
```
